```python
import ternary
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
scale = 100

#figure and axis component
figure, tax = ternary.figure(scale = scale)
figure.set_size_inches(11, 11)
figure.set_facecolor('white')

# Draw Boundary and Gridlines
tax.boundary(linewidth=1.5) #outside triangle boundary width
tax.gridlines(color="black", multiple=10,linewidth = .5) #inside gridline color, total number, linewidth

#make the lines for the feldspar ternary
#cur1x1 = np.linspace(38,37,100)
#cur1y1 = np.linspace(10,11,100)
#cur1z1 = np.linspace(54,55,100)


fields1 = [(90,10,0),(80,10,10),(90,0,10),(80,10,10),(60,10,30),(70,0,30),(60,10,30),(40,10,50),(50,0,50),(40,10,50),
         (38,10,54),(36,10.3,53.7),(34.6,10.7,54.7),(33,11.2,55.9),(31.3,11.8,55.9),(30,12.4,57.6),
           (27.85,13.45,58.5),(25.6,15.1,59.3),(23.4,17,59.6),
           (20,20,60),
           (17,23.4,59.6),(15.1,25.6,59.3),(13.55,27.85,58.4),
          (12.4,30,57.6),(11.8,31.3,55.9),(11.2,33,55.9),(0,36,64),(11.2,33,55.9),(10.7,34.6,54.7),(10.3,36,53.7),
           (10,38,54),(10,40,50),(10,90,0)]
fields2 = [(10,0,90),(10,10,80),(0,10,90),(10,10,80),(11.6,11.4,77),(13,12,75),
           (16.7,13,70),(19.75,13.25,67),(27.85,13.45,58.6),(25.55,13.45,61),(30,0,70)]

# Set Axis labels and Title
fontsize = 20
tax.right_corner_label("An", fontsize=fontsize, offset = 0.2)
tax.top_corner_label("Or", fontsize=fontsize, offset = 0.2)
tax.left_corner_label("Ab", fontsize=fontsize, offset = 0.2)

def tern_points(x,y,z):
    """Tern_points takes 3 equal size 1D arrays and organizes them into points to be plotted on a ternary
        with the following arrangement:(lower right,top,lower left).
            Inputs: 
            x = 1D array like (lower right vertex)
            y = 1D array like (top vertex)
            z = 1D array like (lower left vertex)
            
            Returns:
            points = list of points to be plotted on a ternary diagram 
            utilizing https://github.com/marcharper/python-ternary"""
    points = []
    for i in range(len(x)):
        points.append((x[i],y[i],z[i]))
    return points

#Turning your individual arrays into something that can be plotted on the ternary
#pcorept = tern_points(plincores.An*100,plincores.Or*100,plincores.Ab*100)
#primpt = tern_points(primAn,primOr,primAb)
#rcorept = tern_points(rcoreAn,rcoreOr,rcoreAb)
#rrimpt = tern_points(rrimAn,rrimOr,rrimAb)
#glompt = tern_points(glomsAn,glomsOr,glomsAb)
#acorept = tern_points(acoreAn,acoreOr,acoreAb)
#arimpt = tern_points(arimAn,arimOr,arimAb)

#input feldspar fields
plt.text(84,9.5,'Anorthite',fontsize='12')
plt.text(70,9.5,'Bytownite',fontsize='12')
plt.text(50,9.5,'Labradorite',fontsize='12')
plt.text(35.5,9.5,'Andesine',fontsize='12',rotation =345)
plt.text(15,1,'Oligoclase',fontsize='12')
plt.text(4,3,'Albite',fontsize='12',rotation=25)
plt.text(14,13,'Anorthoclase',fontsize='12')
plt.annotate('Sanidine',xy=(40,60),xytext=(50,55),fontsize='12',arrowprops=dict(arrowstyle='-'))
tax.plot(fields1,linewidth=1,color='k',marker='')
tax.plot(fields2,linewidth=1,color='k')

# Set ticks
tax.ticks(axis='lbr', linewidth=0, multiple=20, offset=0.01,fontsize=12)

# Remove default Matplotlib Axes
tax.clear_matplotlib_ticks()
tax.get_axes().axis('off')
tax._redraw_labels()
tax.legend(frameon = False, ncol=2, labelspacing = 1)
```
